```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import math
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) * 4 - 2
y = 1 - 4 * X + 7 * X**2 + np.random.randn(100, 1) * 3
y = np.round(y, 2)
```

```python
%matplotlib inline
plt.scatter(X, y, color='red', marker='+')
```

## When we try to fit a Line in the given data

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
X_vals = np.linspace(-2, 2, 100).reshape(-1, 1)
y_pred = reg.predict(X_vals)
```

```python
%matplotlib inline
plt.scatter(X, y, color='red', marker='+')
plt.plot(X_vals, y_pred, color='black')
```

## Fitting a Polynomial function in the data

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2)
X_poly_vals = poly_features.fit_transform(X)

poly_reg = LinearRegression()
poly_reg.fit(X_poly_vals, y)
```

```python
X_poly_valsTest = np.linspace(-2, 2, 100).reshape(-1, 1)
X_poly_valsTest_123 = poly_features.transform(X_poly_valsTest)

y_poly_pred = poly_reg.predict(X_poly_valsTest_123)
```

```python
%matplotlib inline 
plt.scatter(X, y, color='red', marker='+')
plt.plot(X_poly_valsTest, y_poly_pred)
```
